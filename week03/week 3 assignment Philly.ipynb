```python
import pandas as pd
```

```python
df=pd.read_csv(
    'data/Philly_Unemployment.csv',
    dtype= 
    {
        'Geo_FIPS':str,
        'Geo_STATE':str,
        'Geo_COUNTY': str
    }
)
```

```python
df.head()
```

```python
df.info(verbose = True,show_counts = True)
```

```python
#dropping and deleting null columns, list of fields below have no data in them 
df.columns[df.isna().all()].tolist()
```

```python
#dropping columns with no data 
df=df.dropna(axis=1,how="all")
```

```python
#check data with dropped null columns 
df.info()
```

```python
#reinspect the data
df.head()
```

```python
#define columns to keep
columns_to_keep=['Geo_FIPS',
                 'SE_A17005_001',
                 'SE_A17005_002',
                 'SE_A17005_003',
                 'PCT_SE_A17005_002',
                 'PCT_SE_A17005_003',]
#add it to a new dataframe
df2=df[columns_to_keep]
```

```python
#look at columns in new dataframe, same as df.columns.to_list
columns=list(df2)
columns
```

```python
df2.columns=['FIPS',
             'Total Labor',
             'Employed',
             'Unemployed',
             'Pct Employed',
             'Pct Unemployed']
```

```python
import geopandas as gpd
```

```python
#read in a gejson file downloaded from the philly open data portal census tract data 
tracts=gpd.read_file('data/Philly_Census_Tracts_2010.geojson')
tracts.head()
```

```python
tracts.plot(figsize=(12,10)) #plot the census tracts 
```

```python
#more about the data 
tracts.info(verbose=True, show_counts=True)
```

```python
#subset the data for FIPS and geometry 
tracts=tracts[['GEOID10','geometry']]
```

```python
tracts.head() #check the data 
```

```python
columns=list(tracts)
columns
```

```python
tracts.columns=['FIPS',
                'geometry']
```

```python
tracts.sample(5)
```

```python
#create a new datafram based on join to merge employment and tract data to create a map 
tracts_labor=tracts.merge(df2,on="FIPS")
```

```python
tracts_labor.head()
```

```python
#simple plot creation with newly joined data 
tracts_labor.plot(figsize=(12,10),
                  column='Unemployed',
                  legend=True,
                  scheme='NaturalBreaks')
```

```python
tracts_labor.plot(figsize=(12,10),
                 column='Unemployed',
                 legend=True, 
                 scheme='equal_interval')
```

```python
tracts_labor.plot(figsize=(12,10),
                 column='Pct Unemployed',
                 legend=True, 
                 scheme='equal_interval')
```

```python
#tracts with 10% or more unemployed - error message idk why 
tracts_labor[tracts_labor.pct unemployed > 10].plot(figsize=(12,10))
```
