```python
import geopandas as gpd
```

```python
#read in a geojson file downloaded from the philly open data portal census tract data 
blocks=gpd.read_file('data/Census_Block_Groups_2010.geojson')
blocks.head()
```

```python
blocks.info()
```

```python
blocks.plot(figsize=(12,10)) #plot the census tracts 
```

```python
#subset the data for FIPS and geometry 
blocks=blocks[['GEOID10','geometry']]
```

```python
blocks.head() #check the data 
```

```python
blocks.columns=['FIPS',
                'geometry']
```

```python
blocks.sample(5)
```

```python
#create a new dataframe based on join to merge employment and tract data to create a map 
blocks_food=blocks.merge(df2,on="FIPS")
```

```python
blocks_food.info()
```

```python
df2.sample(5)
```

```python
blocks_food.head()
```

```python
#simple plot creation with newly joined data 
blocks_food.plot(figsize=(12,10),
                  column='Percent High Produce',
                  legend=True,
                 cmap= 'GnBu',
                  scheme='NaturalBreaks')

plt.title('Percent of High Produce Stores in Philadelphia, 2019')
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# create the map plot
blocks_food.plot(ax=ax,
         column='Percent High Produce',
         cmap='GnBu',
         legend=True,)
         

# add a title
ax.set_title('Percent of High Produce Supply Stores in Philadelphia, 2019')

# get rid of the axis
ax.axis('off');
```

```python
#simple plot creation with newly joined data 
blocks_food.plot(figsize=(12,10),
                  column='Percent High Produce',
                  legend=True,
                 cmap= 'GnBu',
                  scheme='NaturalBreaks')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# create the map plot
blocks_food.plot(ax=ax,
         column='Total High Produce',
         cmap='GnBu',
         legend=True,)
         

# add a title
ax.set_title('Number of High Produce Stores in Philadelphia, 2019')

# get rid of the axis
ax.axis('off');
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# create the map plot
blocks_food.plot(ax=ax,
         column='Total Low Produce',
         cmap='GnBu',
         legend=True,)
         

# add a title
ax.set_title('Number of Low Produce Stores in Philadelphia, 2019')

# get rid of the axis
ax.axis('off');
```

```python
#simple plot creation with newly joined data 
blocks_food.plot(figsize=(12,10),
                  column='Percent High Produce',
                  legend=True,
                 cmap= 'GnBu',
                  scheme='equal_interval')
plt.show()
```
